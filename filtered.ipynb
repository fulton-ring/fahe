# Education

```python
# Path to your Education folder (go one level up first)
folder_path = os.path.join("..", "raw_data", "Education")

# Find all CSV files
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))

dfs = []
for file in csv_files:
    # Extract year from filename (e.g., Education_2019.csv → 2019)
    year = os.path.splitext(os.path.basename(file))[0].split("_")[-1]
    
    df = pd.read_csv(file)
    df["year"] = year   # Add year column
    dfs.append(df)

# Combine all with year included
edu_df = pd.concat(dfs, ignore_index=True)


print(f"Found {len(csv_files)} CSV files:")
for file in csv_files:
    print(f" - {os.path.basename(file)}")


print(edu_df.head())
```

```python
# Delete row at index 0
edu_df = edu_df.drop(0)
# Reset index after deletion
edu_df = edu_df.reset_index(drop=True)
```

```python
# Split NAME column into County and State
edu_df[['County', 'State']] = edu_df['NAME'].str.split(',', expand=True)

# Clean up extra spaces
edu_df['County'] = edu_df['County'].str.strip()
edu_df['State'] = edu_df['State'].str.strip()

# Remove the word "County" from County column
edu_df['County'] = edu_df['County'].str.replace('County', '', regex=False).str.strip()

# Reorder columns: GEO_ID, year, County, State, then the rest
cols = edu_df.columns.tolist()

# Ensure order
new_order = ['GEO_ID','year', 'County', 'State'] + [col for col in cols if col not in ['GEO_ID', 'year', 'County', 'State']]

df_all = edu_df[new_order]
df_all = df_all.drop('NAME', axis=1)
```

```python
# List of columns to keep
keep = [
    "GEO_ID", "NAME", "State", "County", "year",   
    "S1501_C01_059E",   # Median earnings (headline for education table)
]

# Filter to the columns that exist, then rename
existing = [c for c in keep if c in df_all.columns]

# Dictionary for renaming columns
renamed = {
    "S1501_C01_059E": "Total Median Earnings",
}
```

```python
# Create the output DataFrame with existing columns and renamed
df_out = df_all[existing].rename(columns=renamed)
```

```python
df_out.head()
```

```python
# Replace all occurrences of '(X)' with None (null in pandas)
df_out = df_out.replace('(X)', None)
```

```python
df_out.columns = [col.lower().replace(' ', '_') for col in df_out.columns]
```

```python
# Extract only digits from year column, drop bad text
df_out["year"] = df_out["year"].astype(str).str.extract(r"(\d+)")

# Convert to integer
df_out["year"] = df_out["year"].astype(int)

# Build mapping dynamically for 2-digit years
year_map = {y: 2000 + y for y in df_out["year"].unique() if y < 100}

# Apply replacement
df_out["year"] = df_out["year"].replace(year_map)


print(year_map)                # e.g. {15: 2015, 16: 2016}
print(df_out["year"].unique()) # should show full 4-digit years
```

```python
# Read the shapefile using a raw string to avoid invalid escape sequence errors
gdf = gpd.read_file(r'cb_2018_us_county_5m\cb_2018_us_county_5m.shp')
```

```python
from pathlib import Path
```

```python
# Geometry: keep only FIPS + geometry, pad to 5 digits
gdf_slim = gdf[["GEOID", "geometry"]].copy()
gdf_slim["GEOID"] = gdf_slim["GEOID"].astype(str).str.zfill(5)
```

```python
#Prep combined_df: ensure county_fips exists and is 5-digit string
df = combined_df.copy()

if "county_fips" not in df.columns:
    # derive from geo_id/GEO_ID if needed
    if "geo_id" in df.columns:
        df["county_fips"] = df["geo_id"].astype(str).str[-5:]
    elif "GEO_ID" in df.columns:
        df["county_fips"] = df["GEO_ID"].astype(str).str[-5:]
    else:
        raise KeyError("Need 'county_fips' or a 'geo_id'/'GEO_ID' to derive it.")

df["county_fips"] = df["county_fips"].astype(str).str.zfill(5)
```

```python
#Join on county_fips ↔ GEOID
merged = df.merge(
    gdf_slim.rename(columns={"GEOID": "county_fips"}),
    on="county_fips",
    how="left"
)
```
