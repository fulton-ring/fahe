```python
# Calculating percent columns (value / Population Below Poverty) * 100
percent_cols = ["Adults 25 and Older Less Than High School Graduate Percent", "Adults 25 and Older with Bachelor's degree or higher Percent"]
existing_percent_cols = [col for col in percent_cols if col in pov_df.columns]
other_cols = [col for col in pov_df.columns if col not in existing_percent_cols]
pov_df = pov_df[existing_percent_cols + other_cols]


def insert_after(df, after_col, new_col, values):
    cols = list(df.columns)
    if after_col in cols:
        idx = cols.index(after_col) + 1
 
        if new_col in cols:
            cols.remove(new_col)
        cols = cols[:idx] + [new_col] + cols[idx:]
        df[new_col] = values
        df = df[cols]
    else:
        df[new_col] = values
    return df

# Incomplete High School Graduates (Percent)
if "Adults 25 and Older Less Than High School Graduate" in pov_df.columns and "Total Population" in pov_df.columns:
    percent = (
        pd.to_numeric(pov_df["Adults 25 and Older Less Than High School Graduate"], errors="coerce") /
        pd.to_numeric(pov_df["Total Population"], errors="coerce")
    ) * 100
    percent = percent.round(1)
    pov_df = insert_after(
        pov_df,
        "Adults 25 and Older Less Than High School Graduate",
        "Adults 25 and Older Less Than High School Graduate Percent",
        percent
    )

# Bachelor's degree or higher (Percent)
if "Adults 25 and Older with Bachelor's degree or higher" in pov_df.columns and "Total Population" in pov_df.columns:
    percent = (
        pd.to_numeric(pov_df["Adults 25 and Older with Bachelor's degree or higher"], errors="coerce") /
        pd.to_numeric(pov_df["Total Population"], errors="coerce")
    ) * 100
    percent = percent.round(1)
    pov_df = insert_after(
        pov_df,
        "Adults 25 and Older with Bachelor's degree or higher",
        "Adults 25 and Older with Bachelor's degree or higher Percent",
        percent
    )
```

```python
def clean(df):
    df = df.copy()
    df["county_fips"] = df["county_fips"].astype(str).str.zfill(5).str.strip()
    df["year"] = pd.to_numeric(df["year"], errors="coerce").astype("Int64")
    # drop rows with missing merge keys
    df = df.dropna(subset=KEYS)
    # keep the first record per (county_fips, year), no aggregation
    df = df.sort_values(KEYS).drop_duplicates(subset=KEYS, keep="first")
    return df

df_investment = clean(df_investment)
df_poverty    = clean(df_poverty)
df_income     = clean(df_income)
df_education  = clean(df_education)


# Change datatypes of specified columns to int for every df
for df in [df_education, df_income, df_poverty, df_investment]:
    for col in [
        "county_fips",
        "total_investment_dollars",
        "total_number_of_investments",
        "total_population",
        "population_below_poverty",
        "adults_25_and_older_less_than_high_school_graduate",
        "adults_25_and_older_with_bachelor's_degree_or_higher",
        "number_of_households",
        "average_income_per_household",
        "total_median_earnings"
    ]:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors="coerce").astype("Int64")
    # Change population_below_poverty_percent to float 
    if "population_below_poverty_percent" in df.columns:
        df["population_below_poverty_percent"] = pd.to_numeric(df["population_below_poverty_percent"], errors="coerce").astype(float)
```

```python
KEYS = ["county_fips", "year"]
```

```python
base = (df_investment[["county_fips","state","county","year"]]
        .drop_duplicates(subset=KEYS))

final = (
    base
    .merge(df_investment[KEYS + ["total_investment_dollars","total_number_of_investments"]],
           on=KEYS, how="outer", validate="one_to_one")
    .merge(df_poverty[KEYS + ["total_population","population_below_poverty","population_below_poverty_percent",
                              "adults_25_and_older_less_than_high_school_graduate", "adults_25_and_older_less_than_high_school_graduate_percent",
                              "adults_25_and_older_with_bachelor's_degree_or_higher", "adults_25_and_older_with_bachelor's_degree_or_higher_percent"]],
           on=KEYS, how="outer", validate="one_to_one")
    .merge(df_income[KEYS + ["number_of_households","average_income_per_household"]],
           on=KEYS, how="outer", validate="one_to_one")
    .merge(df_education[KEYS + ["total_median_earnings"]],
           on=KEYS, how="outer", validate="one_to_one")
)
```
