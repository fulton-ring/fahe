```python
import pandas as pd
```

```python
import geopandas as gpd
```

```python
# Load the three CSV files
education = pd.read_csv(r"C:\Users\athar\OneDrive\Desktop\fulton_ring\fahe\rough_edu\final_education.csv")
poverty = pd.read_csv(r"C:\Users\athar\OneDrive\Desktop\fulton_ring\fahe\rough_poverty\final_poverty.csv")
aggregator = pd.read_csv(r"C:\Users\athar\OneDrive\Desktop\fulton_ring\fahe\final_data\final_aggregator.csv")
household_income = pd.read_csv(r"C:\Users\athar\OneDrive\Desktop\fulton_ring\fahe\rough\final_household_income.csv")
```

```python
# Concatenate the loaded DataFrames along columns (axis=1)
final_df = pd.concat([education, poverty, aggregator, household_income], axis=1)
```

```python
final_df.head()
```

```python
# Remove duplicate columns (e.g., geo_id, county, state, year, etc.)
final_df = final_df.loc[:, ~final_df.columns.duplicated()]
```

```python
# Define the desired column order
desired_columns = [
    # Only include columns that exist in the DataFrame
    *[col for col in [
        'year',
        'county_fips',
        'county',
        'state_name',
        'total_population',
        'number_of_households',
        'average_income_(per_household)',
        'total_median_earnings',
        'total_502_investment_dollars',
        'total_number_of_502_investment',
        'population_below_poverty',
        'percent_below_poverty',
        'adults(18-24)_with_high_school',
        'percent_adults(18-24)_with_high_school',
        'adults(18-24)_with_college_degree',
        'percent_adults(18-24)_with_college_degree',
        'adults(25+)_with_high_school',
        'percent_adults(25+)_with_high_school',
        'adults(25+)_with_college_degree',
        'percent_adults(25+)_with_college_degree',
        'geo_id',
        'geometry'
    ] if col in final_df.columns]
]

```

```python
from shapely import wkt
```

```python
final_df = final_df[desired_columns].copy()
```

```python
final_df["geometry"] = final_df["geometry"].apply(lambda x: wkt.loads(x) if isinstance(x, str) else x)
```

```python
gdf = gpd.GeoDataFrame(final_df, geometry="geometry", crs="EPSG:4326")
```

```python
gdf.to_file("final_fahe_outcome.geojson", driver="GeoJSON")
```
